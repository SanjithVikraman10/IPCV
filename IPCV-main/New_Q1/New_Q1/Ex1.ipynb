```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image in grayscale
image = cv2.imread('image2.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Step 1: Smooth the image using a Gaussian filter (optional)
smoothed_image = cv2.GaussianBlur(image, (5, 5), 1.5)
```

```python
# Step 2: Compute gradients using Sobel filters
sobel_x = cv2.Sobel(smoothed_image, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in x-direction
sobel_y = cv2.Sobel(smoothed_image, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in y-direction
```

```python
# Step 3: Compute the magnitude of the gradients
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
magnitude = np.uint8(magnitude / np.max(magnitude) * 255)  # Normalize to 0-255
```

```python
# Display the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Sobel X")
plt.imshow(np.abs(sobel_x), cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Sobel Y")
plt.imshow(np.abs(sobel_y), cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Edge Magnitude")
plt.imshow(magnitude, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
